```python
import pandas as pd

# Create the DataFrame
data = {'num': [8, 8, 6, 6, 1, ]}
df = pd.DataFrame(data)

value_counts_df = df['num'].value_counts()
filtered_df = df[df['num'].isin(value_counts_df[value_counts_df == 1].index)]

if not filtered_df.empty:
    max_value = filtered_df.max()
    max_value = pd.DataFrame({'num': max_value})
else:
    max_value = pd.DataFrame({'num': [None]})

display(max_value)
```
